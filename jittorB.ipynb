```python
import shutil
import os

source_directory = '/content/drive/MyDrive/AMU_merge'  # 替换为谷歌云盘中的文件夹路径
target_directory = '/content/AMU_merge'

for root, dirs, files in os.walk(source_directory):
    for file in files:
        if file.endswith(('.py', '.txt', '.yaml')):  # 添加 .txt 和 .yaml 文件类型
            source_file = os.path.join(root, file)
            relative_path = os.path.relpath(root, source_directory)
            target_file = os.path.join(target_directory, relative_path, file)

            os.makedirs(os.path.dirname(target_file), exist_ok=True)
            shutil.copy2(source_file, target_file)
```
